```python
# load dataset
tips = sns.load_dataset("tips")

# Set the figure size
fig1 = plt.figure.Figure(figsize=(14, 8))
ax1 = fig1.add_subplot(1,1,1)

# plot a bar chart
sns.barplot(x="day", y="total_bill", ax=ax1, data=tips, estimator=np.mean, errorbar=('ci', 85), capsize=.2, color='lightblue')
```
